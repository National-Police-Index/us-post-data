```python
import pandas as pd
```

```python
def clean_uid(df):
    df.loc[:, "person_nbr"] = df.person_nbr.str.lower().str.strip().str.replace(r"\s+", "", regex=True)
    return df 

def clean_hire_date(df):
    df.loc[:, "start_date"] = df["Last Hire Date"].astype(str).str.replace(r"(\w+):(\w+):(\w+)", "", regex=True)
    return df.drop(columns=["Last Hire Date"])

def clean_middle_name(df):
    df.loc[:, "middle_name"] = df.middle_name.str.replace(r"\.", "", regex=True)
    return df 

def read_tbl():
    df = pd.read_excel("../../../../data/IN/23-8-2/23-11-2/officer_employment_historymuckrock8172023-20231220144255.xlsx")

    df = df.rename(columns={"Last Name": "last_name", "First Name": "first_name", "Middle Name": "middle_name", 
                            "PSID": "person_nbr", "DOB YEAR": "year_of_birth", 
                            "Employer": "agency_name", 
                            "Title/Rank": "rank", 
                            "Employment Type": "type", 
                            "Appointment Type": "appointment_type",
                            "Employment Status": "separation_reason", })
    return df 


df = read_tbl()

df = df.pipe(clean_hire_date).pipe(clean_uid).pipe(clean_middle_name)

df
```

```python
df.to_csv("../data/output/indiana_index.csv", index=False)
```
