```python
import pandas as pd
```

```python
def clean_middle_name(df):
    df.loc[:, "middle_name"] = df.middle_name.str.replace(r"\.", "", regex=True)
    return df 

def read_tbl():
    df = pd.read_csv("../../../download/data/Ohio/Ohio_index.csv")

    df = df.pipe(clean_middle_name)
    
    df = df.drop(columns=["type"])

    df = df.rename(columns={"agency": "agency_name"})
    return df 


df = read_tbl()

df
```
