```python

# Get the identifying columns
id_cols = ['Person PSID', 'Person First Name', 'Person Middle Name', 'Person Last Name', 'Person Suffix', 'Person Gender']

# Initialize an empty list to store the split rows
split_rows = []

# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    # Iterate over each employment history (0 to 10)
    for i in range(11):
        # Create a new row with identifying columns
        new_row = row[id_cols].tolist()
        
        # Add employment history columns for the current index
        for col in df.columns:
            if col.startswith(f'Employment') and col.endswith(f'_{i}'):
                new_row.append(row[col])
        
        # Append the new row to the split_rows list
        split_rows.append(new_row)

# Get the employment history columns for index 0
emp_cols_0 = [col for col in df.columns if col.startswith('Employment') and col.endswith('_0')]

# Create a new DataFrame from the split rows
new_df = pd.DataFrame(split_rows, columns=id_cols + [col.rsplit('_', 1)[0] for col in emp_cols_0])

new_df
```
