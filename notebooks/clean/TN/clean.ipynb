```python
# drop rows missing start date

new_df
```

```python
newest_df = new_df.rename(columns={"Person PSID": "person_nbr", 
                                "Person First Name": "first_name", 
                                "Person Middle Name": "middle_name", 
                                "Person Last Name": "last_name", 
                                "Employing Organization Name": "agcy_name", 
                                "Employment Change Reason": "type", 
                                "Employment Title/Rank (Current)": "rank", 
                                "Employment Start Date": "start_date", 
                                "Employment End Date": "end_date"})


newest_df = newest_df[["person_nbr", "first_name", "middle_name", "last_name", "agcy_name", "type", "rank", "start_date", "end_date"]]

newest_df.loc[:, "full_name"] = (
    newest_df.first_name.astype(str).fillna("") + " " +
    newest_df.middle_name.astype(str).fillna("") + " " +
    newest_df.last_name.astype(str).fillna("")
).str.strip()

## remove nans in full_name

newest_df
```

```python
# Convert start_date and end_date to datetime format, coercing invalid dates to NaT
newest_df['start_date'] = pd.to_datetime(newest_df['start_date'], errors='coerce')
newest_df['end_date'] = pd.to_datetime(newest_df['end_date'], errors='coerce')

# Extract only the date part
newest_df['start_date'] = newest_df['start_date'].dt.date
newest_df['end_date'] = newest_df['end_date'].dt.date

# Format the dates as 'YYYY-MM-DD', handling NaT values
newest_df['start_date'] = newest_df['start_date'].apply(lambda x: x.strftime('%Y-%m-%d') if not pd.isnull(x) else '')
newest_df['end_date'] = newest_df['end_date'].apply(lambda x: x.strftime('%Y-%m-%d') if not pd.isnull(x) else '')

newest_df
```

```python
# Drop rows where both start_date and end_date are empty strings


### keep rows where start date is an empty string because end date might be present, incidating that the officer once worked at that given agency

## drop asterisks from uid
newest_df.loc[:, "person_nbr"] = newest_df.person_nbr.astype(str).str.replace(r"^\*\*", "", regex=True)


newest_df
```

```python
newest_df.loc[:, "agcy_name"] = (newest_df.agcy_name
                                 .str.lower()
                                 .str.strip()
                                 .str.replace(r"^ (\w+)", r"\1", regex=True)
                                 .str.replace(r"(\w+) $", r"\1", regex=True)
                                 .str.replace(r"(.+)\btbd\b(.+)", "", regex=True)
                                 .str.replace(r"(.+)?fire(.+)?", "", regex=True)
                                 .str.replace(r"(.+) vfd-? ?(.+)?", "", regex=True)
                                 .str.replace(r"dept\.?", "department", regex=True)
                                 .str.replace(r"(\w+) \btn\b (\w+)", r"\1 tennessee \2", regex=True)
                                 .str.replace(r"comm\.? college", "community college", regex=True)
                                 .str.replace(r"^tn ", "tennesee ", regex=True)
                                 .str.replace(r" & ", " and ", regex=False)
                                 .str.replace(r"(.+)\bems\b(.+)", "" ,regex=True)
                                 .str.replace(r"(.+) ems$", "" ,regex=True)
                                 .str.replace(r" tn$", " tennesee", regex=True)
                                )

newest_df
```

```python
unique_agency_names = newest_df['agcy_name'].unique().tolist()

unique_agency_names
```

```python
def clean_type(df):
    df.loc[:, "type"] = df.type.str.replace(r"(\w+) $", r"\1", regex=True)
    return df 



newest_df = newest_df.pipe(clean_type)

newest_df.type.unique()
```
