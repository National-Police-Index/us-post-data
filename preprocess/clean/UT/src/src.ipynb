```python
from datetime import datetime

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../../data/UT/")
```

```python
def convert_date(date_value):
    if pd.isna(date_value):
        return np.nan
    try:
        date_str = str(date_value)
        date_obj = datetime.strptime(date_str, "%m/%d/%y")
        return date_obj.strftime("%Y-%m-%d")
    except ValueError:
        return np.nan


def clean_dates(df):
    df["start_date"] = df.start_date.apply(convert_date)
    df["end_date"] = df.end_date.apply(convert_date)
    return df


def read_tbl():
    df = pd.read_csv("../../../download/data/Utah/Utah_index.csv")

    df = df.rename(
        columns={
            "status": "separation_reason",
            "change_reason": "status",
            "agcy_name": "agency_name",
        }
    )

    df = df.drop(columns=["birth_year", "age", "middle_initial"])
    return df


df = read_tbl().pipe(clean_dates)

df
```

```python
df.to_csv("../data/output/utah_index.csv", index=False)
```
