```python
import pandas as pd
```

```python
def clean_emp_id(df):
    df.loc[:, "emp_id"] = (
        df.emp_id.astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"\s+", "", regex=True)
    )
    return df


def read_tbl():
    df = pd.read_csv("../../../download/data/Alaska/Alaska_index.csv")
    return df


df = read_tbl()

df
```
