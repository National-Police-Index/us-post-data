```python
def clean_uid(df):
    df.loc[:, "person_nbr"] = (
        df.person_nbr.astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"\s+", "", regex=True)
    )
    return df


def clean_sex_and_race(df):
    df.loc[:, "sex"] = df.gender.str.lower().str.strip()

    df.loc[:, "race"] = (
        df.race.str.lower()
        .str.strip()
        .fillna("")
        .str.replace(r"caucasian", "white", regex=True)
        .str.replace(r"no response", "", regex=True)
    )
    return df.drop(columns=["gender"])


def read_tbl():
    dfa = pd.read_csv(
        "../../../download/data/South Carolina/South_Carolina_index.csv"
    )

    dfa = dfa.drop(columns=["suffix", "year_of_birth", "age", "rank"])

    dfa = dfa.pipe(clean_uid)

    dfa = dfa.rename(columns={"agency": "agency_name"})

    # dfb = pd.read_csv("../../../download/data/South Carolina/sc-2023-original-officers.csv")

    # dfb = dfb.pipe(clean_sex_and_race).pipe(clean_uid)

    # dfb = dfb[["person_nbr", "sex", "race"]]

    # df = pd.merge(dfa, dfb, on="person_nbr")
    return dfa


df = read_tbl()


df
```
