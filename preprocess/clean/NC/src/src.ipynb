```python
import pandas as pd
```

```python
def clean_birth_year(df):
    df.loc[:, "year_of_birth"] = df["BirthDate"].astype(str).str.replace(r"^(\w{4})(.+)", r"\1", regex=True)
    return df.drop(columns=["BirthDate"])

def clean_dates(df):
    df.loc[:, "start_date"] = df.start_date.astype(str).str.replace(r"(\w+):(\w+):(\w+)", "", regex=True)
    df.loc[:, "end_date"] = df.end_date.astype(str).str.replace(r"(\w+):(\w+):(\w+)", "", regex=True)
    return df 

def read_tbl():
    df = pd.read_excel("../../../../data/NC/2022-10-10/list_of_all_certified_and_decertified_individuals-20231220144255.xlsx")

    df = df.rename(columns={"IndividualID": "person_nbr", "FirstName": "first_name", "MiddleName": "middle_name", 
                            "LastName": "last_name", "Suffix": "suffix", "AgencyName": "agency_name", 
                            "AppointmentDate": "start_date", 
                            "SeparationDate": "end_date", 
                            "CertificationStatusDescription": "certification_status"})
        
    return df 

def filter_tbl(df):
    df = df[["person_nbr", "first_name", "middle_name", "last_name", "suffix", "agency_name", "start_date", "end_date", "certification_status"]]
    df = df.fillna("")
    return df


df = read_tbl()

df = df.pipe(clean_birth_year).pipe(clean_dates).pipe(filter_tbl)

df.sort_values("person_nbr")
```

```python
df.to_csv("../data/output/north_carolina_index.csv", index=False)
```
