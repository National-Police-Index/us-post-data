```python
import pandas as pd
```

```python
def clean_middle_name(df):
    df.loc[:, "middle_name"] = df.middle_name.str.replace(r"\.", "", regex=True)
    return df


def read_tbl():
    df = pd.read_csv("../../../download/data/Maryland/Maryland_index.csv")

    df = df.rename(columns={"agency": "agency_name"})
    df = df.drop(columns=["year_of_birth", "age", "name"])

    df = df.pipe(clean_middle_name)

    df = df.fillna("")
    return df


df = read_tbl()

df
```

```python
df.to_csv("../data/output/maryland_index.csv", index=False)
```
