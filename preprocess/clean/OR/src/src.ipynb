```python
import pandas as pd
```

```python
def clean_emp_id(df):
    df.loc[:, "emp_id"] = (
        df.emp_id.astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"\s+", "", regex=True)
    )
    return df


def read_tbl():
    dfa = pd.read_csv(
        "../../../download/data/Oregon/or-2023-original-employment.csv"
    )
    dfb = pd.read_csv(
        "../../../download/data/Oregon/or-2023-original-employees.csv"
    )

    dfb = dfb[["emp_id", "sex", "ethnic", "ethnic_description"]]

    dfa = dfa[["emp_id", "action_description"]]

    dfa = dfa.pipe(clean_emp_id)
    dfb = dfb.pipe(clean_emp_id)

    dfc = pd.read_csv("../../../download/data/Oregon/Oregon_index.csv")

    dfc = dfc.rename(columns={"person_nbr": "emp_id"})

    dfc = dfc.pipe(clean_emp_id)

    df = pd.merge(dfa, dfc, on="emp_id")
    df = pd.merge(dfb, df, on="emp_id")
    df = df.rename(
        columns={
            "action_description": "separation_reason",
            "emp_id": "person_nbr",
        }
    )

    df = df.drop(
        columns=["age", "type", "birth_year", "suffix", "middle_name", "ethnic"]
    )
    df = df.rename(
        columns={"middle_initial": "middle_name", "ethnic_description": "race"}
    )
    return df


df = read_tbl()

df
```

```python
df.to_csv("../data/output/oregon_index.csv", index=False)
```
