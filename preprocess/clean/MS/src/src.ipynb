```python
import pandas as pd
import re


def read_tbl():
    df = pd.read_excel("../data/input/ms_response_copyofpostrosterfile.xlsx")

    df = df.rename(columns={"FIRST": "first_name", 
    "LAST": "last_name",
    "STATUS": "employment_status", 
    "DOH": "start_date", 
    "DOT": "end_date",
    "CERT #": "uid"})
    return df 

def filter_uid(df):
    # Create number pattern for 4 or 5 digits
    pattern = r'^\d{4,5}$'
    
    # Filter rows where cert matches pattern
    mask = df['uid'].astype(str).str.match(pattern)
    
    return df[mask]


def fix_agency_name(df):
    df.loc[:, "agency_name"] = df.Agency.str.cat(df.TYPE, sep=" ")
    return df.drop(columns=["Agency", "TYPE"])



def clean_dates(df):
        df['start_date'] = df['start_date'].dt.date
        df['end_date'] = df['end_date'].dt.date
        return df


df = read_tbl()

df = df.pipe(filter_uid).pipe(fix_agency_name).pipe(clean_dates)

df.tail(30)

```
