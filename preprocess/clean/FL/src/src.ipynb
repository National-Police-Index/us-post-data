```python
import pandas as pd
```

```python
def clean_uid(df):
    df.loc[:, "person_nbr"] = (
        df.person_nbr.astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r"\s+", "", regex=True)
    )
    return df


def clean_sex(df):
    df.loc[:, "sex"] = (
        df.sex_code.str.lower()
        .str.strip()
        .str.replace(r"\s+", "", regex=True)
        .str.replace(r"^m$", "male", regex=True)
        .str.replace(r"^f$", "female", regex=True)
    )
    return df.drop(columns=["sex_code"])


def clean_race(df):
    df.loc[:, "race"] = (
        df.race_code.str.lower()
        .str.strip()
        .str.replace(r"^h(.+)?", "hispanic", regex=True)
        .str.replace(r"^a(.+)", "asian", regex=True)
        .str.replace(r"^b(.+)", "black", regex=True)
        .str.replace(r"^w(.+)", "white", regex=True)
        .str.replace(r"o(.+)", "other", regex=True)
        .str.replace(r"^na$", "", regex=True)
    )
    return df.drop(columns=["race_code"])


def read_tbl():
    dfa = pd.read_csv("../../../download/data/Florida/Florida_index.csv")

    dfa = dfa.drop(columns=["suffix", "age", "rank"])

    dfa = dfa.rename(columns={"agency": "agency_name"})
    dfa = dfa.pipe(clean_uid)

    dfb = pd.read_csv(
        "../../../download/data/Florida/fl-2023-original-officers.csv"
    )

    dfb = dfb.pipe(clean_uid).pipe(clean_sex).pipe(clean_race)

    dfb = dfb[["race", "sex", "person_nbr"]]

    df = pd.merge(dfa, dfb, on="person_nbr")

    return dfa, dfb, df


dfa, dfb, df = read_tbl()

df
```
